```python
stats_Population_moyenne = stats.Population.mean()
def remean_population(row):
    row.Population = row.Population - stats_Population_moyenne
    return row

stats.apply(remean_population, axis='columns')
stats.head()
```

```python
stats.groupby('Continent').Continent.count()
```
